```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = "https://m.blog.naver.com/ggilhee/223396399295"
respones = requests.get(url)
```

```python
bs_result = BeautifulSoup(respones.text, "html.parser")
bs_result
```

```python
#원하는 정보 가져오기
result = bs_result.find_all("span", class_ = "se-fs-" )
result_list = []
for i in result:
    result_list.append(i.text)
print(result_list)
```
