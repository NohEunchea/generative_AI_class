```python
# 2. html 요소 얻기
url = "https://namu.wiki/w/%EC%97%BC%EB%AC%B4%EC%99%95"
page = requests.get(url)
```

```python
bs_result = BeautifulSoup(page.text, "html.parser")
bs_result
```
