오리 이미지 - numpy 맛보기

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 오리 이미지
duck_img = Image.open("오리이미지2.webp")
img_array = np.array(duck_img)
print(img_array)
plt.imshow(duck_img)
```

```python
img_array_copy1 = img_array.copy()
img_array_copy1[:, :, 0] = 0
plt.imshow(img_array_copy1)
```

```python
img_array_copy2 = img_array.copy()
img_array_copy2[:,:,1] = 0
print(img_array_copy2)
plt.imshow(img_array_copy2)
```

```python
img_array_copy3 = img_array.copy()
plt.imshow(img_array_copy3[200:-200, 200:-200])
```

numpy 연산

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,7,10])

print("합: ", arr1+arr2)
print("차:", arr1-arr2)
print("스칼라곱:", 2*arr1)
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[7,8,9],[10,11,12]])
print(arr1*arr2)
print(arr1/arr2)
```

선형대수 연산

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

arr3 = arr1 @ arr2
print("행렬곱셈:", arr3)

arr1_inv = np.linalg.inv(arr1)
print("역행렬:", arr1_inv)
```

통계

```python
arr = np.array([[1,2,3],[4,5,6]])
print("전체합:", arr.sum())
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print("열의 합:", arr.sum(axis=0))
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print("행의 합:", arr.sum(axis=1))
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print("평균:", arr.mean())
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print("분산:", arr.var())
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print("표준편차:", arr.std())
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print("최댓값:", arr.max())
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print("최솟값:", arr.min())
```
